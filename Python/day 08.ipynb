```python
# create a sample DataFrame with NaN values
df = pd.DataFrame({'A': [1, 2, np.nan, 4, 9],
                   'B': [np.nan, 2, 3, np.nan, 5],
                   'C': [1, 2, 3, 4, np.nan]})

# print the original DataFrame
print(df)

# interpolate the NaN values using the default method (linear interpolation)
df_interpolated = df.interpolate() 
# The interpolate will assume the number and it will automaticallly fill
#if the value at the begining it can't assume because it is at start.
# If the value is at the end it will print the previous value

# print the interpolated DataFrame
print(df_interpolated)
```

```python
#backwardfill
#fordward fill
# create a sample dataframe with NaN values
df = pd.DataFrame({'A': [1, 2, np.nan, 4],
                   'B': [5, np.nan, 7, 8],
                   'C': [9, 10, 11, np.nan]})
print(df)

# fill NaN values using ffill method
df_ffill = df.ffill()
print(df_ffill)

# fill NaN values using bfill method
df_bfill = df.bfill()
print(df_bfill)
```

```python
import pandas as pd

# Sample data with a categorical column
df = pd.DataFrame({
    'Color': ['Red', 'Blue', 'Green', 'Blue', 'Red'],
    'Size': ['S', 'M', 'L', 'S', 'M']
})

print("Original DataFrame:")
print(df)

# One-hot encoding the 'Color' column
df_dummies = pd.get_dummies(df, columns=['Color'], drop_first=True)
print("\nDataFrame with get_dummies applied:")
print(df_dummies)
```

```python


import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(500,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python


# Scatter plot
my_df.plot(kind="scatter", x="Mon", y="Tues", c="Wed")
```

```python


# Colors: https://matplotlib.org/stable/tutorials/colors/colormaps.html
my_df.plot(kind="scatter", x="Mon", y="Tues", c="Wed", cmap='magma')
     
```

```python

# By Size
my_df.plot(kind="scatter", x="Mon", y="Tues", s=abs(my_df['Mon']*100))
```

```python

# By Size easier to read with Alpha
my_df.plot(kind="scatter", x="Mon", y="Tues", s=my_df['Tues']*200, alpha=0.4)
```

```python


# Boxplots
my_df.plot(kind="box")
```

```python


# Boxplots Size
my_df.plot(kind="box", figsize=(10, 5))
```

```python


# Boxplots Color
my_df.plot(kind="box", cmap="jet")
```

```python

# No Legend
my_df.plot(kind="box", lengend=True)
```

```python


# Title
my_df.plot(kind="box", title="My Awesome Box Plot!")
```

```python
# Add Grid
my_df.plot.box(title="My Awesome Box Plot!!!!!!!!", grid=True)
```

```python

# Hexplots
my_df.plot(kind="hexbin", x="Mon", y="Tues")
     
```

```python
# Hexplots gridsize
my_df.plot(kind="hexbin", x="Mon", y="Tues", gridsize=10)
```

```python


# Grid
my_df.plot.hexbin(x="Mon", y="Tues", C="Wed", gridsize=10, title="My Awesome Hex Plot", grid=True)
```
